```python
# 
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu = 10 
N = 10000
y = stats.expon.rvs(scale=mu, size=N)
plt.hist(y, 60, density=True) 
t = np.linspace(0, 100, N // 10) 
pdf = stats.expon.pdf(t, scale=mu) 
plt.plot(t, pdf, "red") 
plt.show()
```

```python
mu = 2
M = 1000 
x = stats.expon.rvs(scale=mu, size=M) 
plt.plot(np.ones(M) * mu, "r.-")
for k in range(1, M + 1): 
    plt.plot(k, np.mean(x[:k]**2), "k.")
plt.legend(["Sant $\mu$", "Skattning av $\mu$."]) 
plt.show()
```

```python
mu = 2
N = 10000 
y = stats.expon.rvs(scale=mu, size=N) 
print(np.mean(y))
```

```python
# Monte carlo pi
size_ = 100000
U = stats.uniform.rvs(size=size_)
V = stats.uniform.rvs(size=size_)
plt.plot(U,V, 'o')

Z = np.sqrt(U**2 + V** 2 <= 1)
pi = np.mean(4 *Z)
print(pi)
```

```python
lambda_ = 0.1
P_3 = stats.expon.cdf(3, scale=1/lambda_)
print(P_3)


P_7 = stats.expon.cdf(7, scale=1/lambda_)
print(1-P_7)

P_3 = stats.expon.cdf(3, scale = 1/lambda_) 
P_4 = stats.expon.cdf(4, scale = 1/lambda_) 

print(P_4-P_3)
```
